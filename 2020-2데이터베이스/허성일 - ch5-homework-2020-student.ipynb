#### 1.  관계형 데이터베이스 테이블 스키마(create table)을 생성하고, MySQL DBMS를 이용하여 테이블을 생성하시오. 

- 테이블은 적절한 컬럼 데이터 타입, Primary Key, Foreign Key (on delete, on update rule 포함, restrict, set null, cascade), 필요하다면 Unique, NOT NULL 등이 표시되어야 한다. 

- 조건: 선수과목 정보는 course 정보 때문에 유지하는 정보이다. 즉, Course가 삭제되면, 불필요한 정보이다.
- 스키마 작성시 9번 문제가 성공할 수 있도록 스키마를 작성하여야 한다.

- Semester은 enum type을 사용하여, 봄, 여름, 가을, 겨울 순서가 되게 한다.
- Grade는 enum type을 사용하여 A, B, C, D, F 순서가 되게 한다.

````python
%%sql
-- YOUR CODE HERE 20

CREATE TABLE STUDENT
(
        Name VARCHAR(15) NOT NULL,
        Student_number INT NOT NULL,
        Class INT NOT NULL,
        Major VARCHAR(8) NOT NULL,
        PRIMARY KEY (Student_number)
) ENGINE=InnoDB;

CREATE TABLE COURSE
(
        Course_name VARCHAR(30) NOT NULL,
        Course_number VARCHAR(12) NOT NULL,
        Credit_hours INT NOT NULL,
        Department VARCHAR(8) NOT NULL,
        PRIMARY KEY (Course_number),
        UNIQUE (Course_name)
) ENGINE=InnoDB;

CREATE TABLE SECTION
(
        Section_identifier INT  NOT NULL,
        Course_number VARCHAR(12) NOT NULL,
        Semester ENUM('Spring','Summer','Fall','Winter'),
        Year CHAR(2)  NOT NULL,
        Instructor VARCHAR(10) NOT NULL,
        PRIMARY KEY (Section_identifier),
        FOREIGN KEY (Course_number) 
                REFERENCES COURSE(Course_number)
                ON DELETE RESTRICT
                ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE PREPEQUISITE
(       
        Course_number VARCHAR(10) NOT NULL,
        Prerequisite_number VARCHAR(10) NOT NULL,
        PRIMARY KEY (Course_number,Prerequisite_number),
        FOREIGN KEY (Course_number) 
                REFERENCES COURSE(Course_number)
                ON DELETE CASCADE
                ON UPDATE CASCADE,
        FOREIGN KEY (Prerequisite_number) 
                REFERENCES COURSE(Course_number)
                ON DELETE CASCADE
                ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE GRADE_REPORT(
        Student_number INT NOT NULL,
        Section_identifier INT NOT NULL,
        Grade ENUM('A','B','C','D','F'),
        PRIMARY KEY (Student_number,Section_identifier),
        FOREIGN KEY (Student_number) 
                REFERENCES STUDENT(Student_number)
                ON DELETE CASCADE
                ON UPDATE CASCADE,
        FOREIGN KEY (Section_identifier) 
                REFERENCES SECTION(Section_identifier)
                ON DELETE RESTRICT
                ON UPDATE CASCADE
) ENGINE=InnoDB;
````

#### 5. Database 과목을 듣기 위해 (직접 또는 간접적으로: directly or indirectly) 미리 들어야 하는(prerequisite) 과목(Course_name)들을 모두 나열하시오. (Prerequisite는 Acyclic graph이고 path의 최대 길이는 3이라고 가정함. 길이 3: A-B-C-D)

````python
%%sql
-- YOUR CODE HERE 10
(
    SELECT cc.Course_name
    FROM COURSE c, PREPEQUISITE p, COURSE cc
    WHERE c.Course_name='Database' and c.Course_number=p.Course_number and cc.Course_number=p.Prerequisite_number
)
UNION ALL
(
    SELECT cc.Course_name
    FROM COURSE c, PREPEQUISITE p, PREPEQUISITE pp, COURSE cc
    WHERE c.Course_name='Database' and c.Course_number=p.Course_number and p.Prerequisite_number=pp.Course_number and cc.Course_number=pp.Prerequisite_number
)
UNION ALL
(
    SELECT cc.Course_name
    FROM COURSE c, PREPEQUISITE p, PREPEQUISITE pp, PREPEQUISITE ppp, COURSE cc
    WHERE c.Course_name='Database' and c.Course_number=p.Course_number and p.Prerequisite_number=pp.Course_number and pp.Prerequisite_number=ppp.Course_number and cc.Course_number=ppp.Prerequisite_number
);
````

#### 6. 전공으로 수강한 과목들만 나열하는 MAJOR_GRADE_REPORT테이블을 만드시오. 테이블을 생성하고, GRADE_REPORT 테이블에서 전공으로 수강한 경우에 해당하는 행들만 아래 테이블에 삽입하시오. 전공으로 수강하였다는 것은 학생의 전공(Major)과 COURSE의 개설 Department가 같으면 전공으로 수강한 과목이다.
   
    MAJOR_GRADE_REPORT(Student_number, Section_identifier, Grade)

````python
%%sql
-- YOUR CODE HERE 10

DROP TABLE IF EXISTS MAJOR_GRADE_REPORT;

CREATE TABLE MAJOR_GRADE_REPORT(
        Student_number INT NOT NULL,
        Section_identifier INT NOT NULL,
        Grade ENUM('A','B','C','D','F'),
        PRIMARY KEY (Student_number,Section_identifier),
        FOREIGN KEY (Student_number) 
                REFERENCES STUDENT(Student_number)
                ON DELETE CASCADE
                ON UPDATE CASCADE,
        FOREIGN KEY (Section_identifier) 
                REFERENCES SECTION(Section_identifier)
                ON DELETE RESTRICT
                ON UPDATE CASCADE
) ENGINE=InnoDB;

INSERT INTO MAJOR_GRADE_REPORT (Student_number, Section_identifier, Grade)
SELECT g.Student_number, g.Section_identifier, g.Grade
FROM GRADE_REPORT g, SECTION s, COURSE c, STUDENT e
WHERE g.Section_identifier=s.Section_identifier and s.Course_number=c.Course_number and g.Student_number=e.Student_number and c.Department=e.Major;
````

#### 7. 모든 학생들의 성적들을 다음과 같이 출력하려 한다. 적절한 SQL문을 작성하여 실행하시오. 
- 결과 테이블은 Major, Student_number, Year, Semester, Course_number 순서로 정렬되어야 한다.
- 결과 테이블 애트리뷰트 이름 및 순서: (Major, Student_number, Student_name, Year, Semester, Course_number, Course_name, Grade)

````python
%%sql
-- YOUR CODE HERE 10

SELECT e.Major, e.Student_number, e.Name, s.Year, s.Semester, s.Course_number, c.Course_name, g.Grade
FROM STUDENT e, GRADE_REPORT g, SECTION s, COURSE c
WHERE c.Course_number=s.Course_number and s.Section_identifier=g.Section_identifier and g.Student_number=e.Student_number
ORDER BY e.Major, e.Student_number, s.Year, s.Semester, s.Course_number;
````
